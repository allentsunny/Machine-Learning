```python
import pandas as pd
df=pd.read_csv('BankCustomerData.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
cols=['job','marital','education','default','housing','loan','contact','month','poutcome','term_deposit']
le=LabelEncoder()
for col in cols:
    df[col]=le.fit_transform(df[col])
df.head()
```

```python
x=df.drop(['loan'],axis=1)
y=df['loan']
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
sl=StandardScaler()
X_train=sl.fit_transform(X_train)
X_test=sl.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
cls=RandomForestClassifier(n_estimators=10,criterion="entropy")
cls.fit(X_train,y_train)
```

```python
y_pred=cls.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
