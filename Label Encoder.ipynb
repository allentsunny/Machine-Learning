```python
import pandas as pd
# one_hot_encoding
df=pd.DataFrame({'Fruits':['apple','banana','orange','kiwi','pineapple']})
print('DataFrame before encoding')
print(df['Fruits'])
one_hot=pd.get_dummies(df['Fruits'])
print('DataFrame after encoding')
one_hot
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
print('Befor encoding')
print(df['Fruits'])
df['fruits_label']=le.fit_transform(df['Fruits'])
print('After encoding')
df
```
