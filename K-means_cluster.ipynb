```python
new_df=df.copy()
new_df
```

```python
x=new_df[['Annual Income (k$)','Spending Score (1-100)']]
x.head()
```

```python
wcss=[]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
import matplotlib.pyplot as plt
plt.plot(wcss,marker='x')
plt.xlabel('x values')
plt.ylabrl('wcss')
plt.show
```

```python
wcss
```

```python
km2 = KMeans(n_clusters=4)
km2
```

```python
km2.fit(x)
```

```python
y_pred=km2.predict(x)
y_pred
```

```python
new_df['Output']=y_pred
new_df.head()
```

```python
sns.scatterplot(x=new_df['Annual Income (k$)'],y=new_df['Spending Score (1-100)'],hue=new_df['Output'])
```

```python
from sklearn import metrics
score = metrics.silhouette_score(x,new_df["Output"])
score
```
